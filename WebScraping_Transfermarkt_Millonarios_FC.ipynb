```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL del sitio web que queremos scrapear
url = 'https://www.transfermarkt.co/millonarios-fc/spielplan/verein/2350/saison_id/2023'

# Hacer una solicitud GET a la página web con un encabezado de agente de usuario
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
response = requests.get(url, headers=headers)

# Verificar que la solicitud fue exitosa (código 200)
if response.status_code == 200:
    # Crear una instancia de BeautifulSoup con el contenido HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encontrar todas las tablas en la página
    tables = soup.find_all('table')

    # Lista para almacenar los DataFrames de cada tabla
    dfs = []

    # Iterar sobre cada tabla y convertirla en un DataFrame
    for index, table in enumerate(tables):
        # Leer la tabla HTML en un DataFrame
        df = pd.read_html(str(table))[0]
        dfs.append(df)

        # Opcional: guardar cada DataFrame en un archivo CSV separado
        df.to_csv(f'millonarios_fc_2023_table_{index+1}.csv', index=False)

    # Mostrar cada DataFrame
    for index, df in enumerate(dfs):
        print(f'Tabla {index+1}:')
        print(df)
else:
    print(f'Error: {response.status_code}')
```

```python
# Cargamos los datos de la tabla 5 que contiene los partidos de todos contra todos
df_TcT = pd.read_csv('millonarios_fc_2023_table_5.csv')
df_TcT
```

```python
# Cargamos los datos de la tabla 7 que contiene los partidos de los cuadrangulares
partidos_C = pd.read_csv('millonarios_fc_2023_table_7.csv')
partidos_C.head()
```

```python
# Unimos los dos DataFrames
df_unido = pd.concat([df_TcT, partidos_C], ignore_index=True)
df_unido.head(23)
```

```python
# Cambiar los valores de la columna 'Jornada' en las filas
df_unido.loc[19:23, 'Jornada'] = [20,21,22,23,24]
df_unido.tail()
```

```python
# Eliminamos la ultima fila ya que no hay información todavia por que no se ha jugado el partido

df_unido = df_unido.drop([24])
df_unido
```

```python
# Expresión regular para extraer el nombre y la posición
pattern = r'(?P<Nombre>.+?)\s?\((?P<Posicion>\d+)\)?'

# Extraemos la posición y el nombre utilizando la expresión regular
df_unido[['Adversario', 'Posicion Adversario']] = df_unido['Adversario.1'].str.extract(pattern)

# Convertir la columna 'Posicion' a enteros, llenando los NaN con un valor por defecto, por ejemplo, 0
df_unido['Posicion Adversario'] = df_unido['Posicion Adversario'].fillna(0).astype(int)

# Mostrar el DataFrame resultante
df_unido
```

```python
# Cambiar los valores de la columna 'Puesto' en las filas
df_unido.loc[0:23, 'Puesto'] = [6,1,4,6,11,4,3,6,10,13,15,16,16,16,12,9,9,7,6,4,4,4,4,4]
df_unido.tail()
```

```python
# Eliminamos la columna 'Adversario'
df_unido.drop(columns=['Adversario.1'], inplace=True)
df_unido.head()
```

```python
# Cambiamos 'H' a '0' y 'A' a '1' en la columna 'Localidad'
df_unido['Localidad'] = df_unido['Localidad'].map({'H':'0', 'A':'1'})
df_unido
```

```python
# Eliminamos la columna 'Horario'
df_unido.drop(columns=['Horario'], inplace=True)
df_unido.head()
```

```python
# Creamos dos nuevas columnas 'goles local' y 'goles visitantes' apartir de la columna 'Resultado'
df_unido[['goles local', 'goles visitante']] = df_unido['Resultado'].str.split(':', expand=True).astype(int)
df_unido
```

```python
# Eliminamos la columna 'Resultado'
df_unido.drop(columns=['Resultado'], inplace=True)
df_unido
```

```python
# Información General
df_unido.info()
```

```python
import re

df_unido['Fecha'] = df_unido['Fecha'].apply(lambda fecha: re.search(r"\d{2}/\d{2}/\d{4}", fecha).group() if re.search(r"\d{2}/\d{2}/\d{4}", fecha) else None)

# Verificar la columna 'Fecha'
print(df_unido['Fecha'])
```

```python
# Convertir los tipos de datos
df_unido['Jornada'] = df_unido['Jornada'].astype(int)
df_unido['Fecha'] = pd.to_datetime(df_unido['Fecha'], format='%d/%m/%Y')
df_unido['Puesto'] = df_unido['Puesto'].astype(int)
df_unido['Localidad'] = df_unido['Localidad'].astype(int)
df_unido['Adversario'] = df_unido['Adversario'].astype('category')
df_unido['Sistema de juego'] = df_unido['Sistema de juego'].astype('category')
df_unido['espectadores'] = df_unido['espectadores'].astype(float)
```

```python
df_unido.info()
```

```python
# Verificar valores nulos
df_unido.isnull().sum()
```

```python
# Reemplazamos los valores NaN por la media en la columna 'espectadores'
df_unido['espectadores'].fillna(df_unido['espectadores'].mean(), inplace=True)
# Verificamos reusltados
print(df_unido['espectadores'].isnull().sum())
```

```python
# Asignamos las posiciones desde la fila 19 en adelante
df_unido.loc[0:23, 'Posicion Adversario'] = [9, 19, 16, 3, 7, 5, 14, 20, 14, 2, 11, 13, 10, 11, 3, 17, 7, 4, 16, 1, 2, 3, 3, 2]
df_unido
```

```python
df_unido.info()
```

```python
# Aplicamos One-hot enconding en 'Sistema de Juego'
df_unido = pd.get_dummies(df_unido, columns=['Sistema de juego'], drop_first=True)
df_unido.info()
```
